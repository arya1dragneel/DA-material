### Data-types

#### Numeric

- Integer (int)
- Float (float)
- Complex (complex)

```python
number1 = 27
number2 = 31.5476
number3 = 67+21j
```

##### type():

```python
print(type(number1))
print(type(number2))
print(type(number3))
```

```python
a = -327
b = -401.3245
c = 8j

print(type(a))
print(type(b))
print(type(c))
```

```python
print('Number',a,'is of type',type(a))
print('Number',b,'is of type',type(b))
print('Number',c,'is of type',type(c))
```

#### Casting Method:

This method is used to convert one data type into another.

```python
p = 23 #this is integer
print(p)
```

```python
p = float(p)
print(p)
type(p) # Now this is float.
```

```python
q = 54
print(q)
type(q)
```

```python
q = complex(q)
print(q)
type(q)
```

```python
# float to int and complex
s = 34.10
print(s)
type(s)
```

```python
s = int(45.23)
print(s)
type(s)
```

```python
w = 43.21
print(w)
type(w)
```

```python
x = complex(w)
print(x)
type(x)
```

**Notes:**

- If you want to convert into integer then use **int()**

- If you want to convert into float then use **float()**

- If you want to convert into complex then use **complex()**
