#### String Data Type (str)

Anything you write inside quotes is termed to be of string data type.

```python
text = 'My name is Shubham.'
print(text)
print('-'*20)
print('I am learning Python.')
```

```python
type(text)
```

```python
s = '34.3210+36j'
d = "Jalgaon "
print(type(s))
print(type(d))
```

### String Methods:

##### strip():

```python
a = '       Hello,                    World !                '
a
```

```python
# strip - whitespaces/blanks removal method at start and at end.
a.strip()
```

##### lower():

```python
text = 'Shubham'
text.lower()
```

##### upper():

```python
text = 'PytHon'
text.upper()
```

##### replace():

```python
text = 'Hello Everyone.'
text.replace('e','5.22')
```

##### split():

```python
text = 'Good Afternoon 2'
print(text.split('2'))
```

##### concatenation:

```python
a='Hello'
b='I am good.'
c=2.4357


print(a,b,c)
print('\n')
print(a+' '+b)
```

##### strip:

```python
# Leading and trailing whitespaces are removed

string = '  xoxo love xoxo   '

print(string.strip())
```
